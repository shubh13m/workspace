```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('titanic_data.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
print(dataset['Age'])
```

```python
dataset['Age'] = dataset['Age'].fillna(int(dataset['Age'].mean()))
```

```python
dataset['Age'].isnull().sum()
```

```python
dataset = dataset.drop(['Cabin', 'Name', 'Ticket', 'Embarked'], axis=1)
```

```python
dataset = pd.concat([dataset, pd.get_dummies(dataset['Sex'])], axis=1)
```

```python
dataset = dataset.drop(['Sex', 'male'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
dataset.isnull().sum()
```

```python
x_train,  x_test, y_train, y_test = train_test_split(dataset.drop('Survived', axis=1), dataset['Survived'], test_size=0.30, random_state=70)
```

```python
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
```

```python
y_predict = nb_model.predict(x_test)
```

```python
nb_model.score(x_test, y_test)
```

```python
for each in range(0,len(y_predict)):
    if(y_predict[each] != y_test.values[each]):
        print(x_test.values[each])
```
