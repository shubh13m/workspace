```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('income.csv')
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
len(dataset)
```

```python
plt.plot(dataset['Age'], dataset['Income($)'], '*', color='red')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(dataset[['Age', 'Income($)']])
```

```python
dataset['cluster'] = y_predicted
```

```python
dataset
```

```python
data1 = dataset[dataset['cluster'] == 0]
data2 = dataset[dataset['cluster'] == 1]
data3 = dataset[dataset['cluster'] == 2]
```

```python
plt.plot(data1['Age'], data1['Income($)'], '+', color='red')
plt.plot(data2['Age'], data2['Income($)'], '+', color='blue')
plt.plot(data3['Age'], data3['Income($)'], '+', color='green')
```

```python
scaler  = MinMaxScaler()
scaler.fit(dataset[['Age']])
dataset['Age'] = scaler.transform(dataset[['Age']])
scaler.fit(dataset[['Income($)']])
dataset['Income($)'] = scaler.transform(dataset[['Income($)']])
```

```python
y_predicted = km.fit_predict(dataset[['Age', 'Income($)']])
dataset['cluster'] = y_predicted
```

```python
data1 = dataset[dataset['cluster'] == 0]
data2 = dataset[dataset['cluster'] == 1]
data3 = dataset[dataset['cluster'] == 2]
```

```python
plt.plot(data1['Age'], data1['Income($)'], '+', color='red')
plt.plot(data2['Age'], data2['Income($)'], '+', color='blue')
plt.plot(data3['Age'], data3['Income($)'], '+', color='green')
```
